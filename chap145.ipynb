#### Chap 1.4.5 Morphological Transformations

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

##### 1.Erosion

```python
img = cv2.imread('j.jpg',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('j.jpg')
plt.subplot(122),plt.imshow(erosion),plt.title('j.jpg')
```

##### 2. Dilation

```python
img = cv2.imread('j.jpg',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('j.jpg')
plt.subplot(122),plt.imshow(dilation),plt.title('j.jpg')
```

##### 3. Opening

```python
img = cv2.imread('jj.jpg',0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('jj.jpg')
plt.subplot(122),plt.imshow(opening),plt.title('jj.jpg')
```

##### 4. Closing

```python
img = cv2.imread('jjj.jpg',0)
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('jjj.jpg')
plt.subplot(122),plt.imshow(closing),plt.title('jjj.jpg')
```

##### 6. Top Hat

```python
img = cv2.imread('j.jpg',0)
kernel = np.ones((10,10),np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('j.jpg')
plt.subplot(122),plt.imshow(tophat),plt.title('j.jpg')
```

##### 7. Black Hat


```python
img = cv2.imread('j.jpg',0)
kernel = np.ones((10,10),np.uint8)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('j.jpg')
plt.subplot(122),plt.imshow(blackhat),plt.title('j.jpg')
```
