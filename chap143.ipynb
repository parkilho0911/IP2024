#### Chap 1.4.3 Geomtric Transform

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('hero.jpg')
res = cv2.resize(img,None,fx=5, fy=5, interpolation = cv2.INTER_CUBIC)

```

```python
cv2.namedWindow('image')
while(1):
    cv2.imshow('image',res)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
```

```python
img = cv2.imread('hero.jpg')
rows,cols = img.shape[:2]
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Rotate

```python
img = cv2.imread('lina.jpg',0)
rows,cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/4,rows/4),45,1)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Affine Transformation


```python
img = cv2.imread('lina.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```

##### Perspective Transformation

```python
img = cv2.imread('chang.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
